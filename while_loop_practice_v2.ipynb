Q4. Find sum of numbers from 1 to n.

Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
#write answer here
num = int(input("Enter number to reverse: "))
original = num
reversed_num = 0

print(f"Original: {original}")

while num > 0:
    last_digit = num % 10
    print(f"Last digit: {last_digit}")

    reversed_num = (reversed_num * 10) + last_digit
    print(f"Reversed so far: {reversed_num}")

    num = num // 10
    print(f"Remaining num: {num}\n")

print(f"Reversed number: {reversed_num}")
```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
#write answer here
num = int(input("Enter number: "))
original = num
count = 0

print(f"Original: {original}")

while num > 0:
    num = num // 10
    count = count + 1
    print(f"After //10: {num}, Count: {count}")

print(f"Number of digits in {original}: {count}")
```

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
#write answer here
num = int(input("Enter number to reverse: "))
original = num
reversed_num = 0

print(f"Original: {original}")

while num > 0:
    last_digit = num % 10

    reversed_num = (reversed_num * 10) + last_digit

    num = num // 10

print(f"Reversed number: {reversed_num}")
if original == reversed_num:
  print("Palindrome")
else:
  print("Not Palindrome")
```

Q8. Guess the number game.

Approach:
- Use while loop
- Continue until correct guess

```python
#write answer here
guess_number = 10
user_guess = int(input("Enter your guess number"))
while user_guess != guess_number:
  if guess_number > user_guess:
    print("Too Low")
  else:
    print("Too High")
  user_guess = int(input("Enter your guess number"))
print("Correct Guess")
```
