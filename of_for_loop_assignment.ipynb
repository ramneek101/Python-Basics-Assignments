## Question 9 (Easy)

**Title:** Count Even and Odd Numbers

### Problem Statement:
Take N as input and count how many even and odd numbers are there between 1 and N.

### Approach:
Use two counters and check with modulo operator.


```python
# write your code here

n = int(input("Enter N: "))

even_count = 0
odd_count = 0

print(f"Counting even/odd from 1 to {n}:")

for i in range(1, n + 1):
    if i % 2 == 0:
        even_count = even_count + 1
        print(even_count)
    else:
        odd_count = odd_count + 1
        print(odd_count)

print(f"Even numbers: {even_count}")
print(f"Odd numbers: {odd_count}")
print(f"Total: {even_count + odd_count}")
```

## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
# write your code here
num = int(input("Enter number to check: "))
if num <= 1:
    print(f"{num} is NOT prime (must be > 1)")
elif num == 2:
    print("2 is PRIME ")
else:
    is_prime = True
    for i in range(2, num):
        print(f"Check {i}: {num} % {i} = {num % i}")
        if num % i == 0:
            is_prime = False
            print(f"{num} divisible by {i} → NOT PRIME")
            break
    if is_prime:
        print(f"{num} is PRIME")
    else:
        print(f"{num} is NOT prime")
```

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
# write your code here
n = int(input("Enter N: "))

print(f"Prime numbers between 1 and {n}:")
for num in range(2, n + 1):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
```

## Question 18 (Hard)

**Title:** First Number Divisible by 7 and 11

### Problem Statement:
Take N as input and find the first number between 1 and N that is divisible by both 7 and 11. Stop after finding it.

### Approach:
Use logical AND and stop the loop using `break`.


```python
# write your code here
n = int(input("Enter N: "))
for i in range(1, n + 1):
    if (i % 7 == 0) and (i % 11 == 0):
        print(f"Found: {i} (divisible by 7 AND 11)")
        break
    else:
        print(f"{i}: 7?{i%7==0} 11?{i%11==0} → No")
```
